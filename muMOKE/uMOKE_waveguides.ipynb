```python
import mumokepy as mumky
from importlib import reload
import matplotlib.pyplot as plt
import numpy as np
reload(mumky)
folder = r'C:\Users\jc2713\OneDrive - Imperial College London (1)\Tohoku - pump priming\Data\muMOKE\No3-Co1Pt2x5-nocap'

files = [r'nocap-STFMR-waveguide-hystloop-Al',
         r'nocap-STFMR-waveguide-hystloop-SiO2',
         r'nocap-ThinFilmHystLoop']

nominal_sizes = np.linspace(300, 800, 9)

arrays = {}
fig_coer,ax_coer = plt.subplots(1,1)
for i, file in enumerate(files):
    arrays[i] = mumky.MOKEAnalyzer(material= 'CoPt5x', sample_name= 'J-wg-3', array_name = 'Film'+str(i), filename = file+'.csv', folder=folder)
    arrays[i].add_property('nominal_diameter', nominal_sizes[i])
    arrays[i].read_data()
    #print(arrays[i].data)
    arrays[i].analyze(field_threshold = 2000)
    fig,ax = plt.subplots(1,1)
    arrays[i].plot_data(ax)




```

```python
coer = []
for arr in list(arrays.values()):
    coer.append(arr.coercive_field)
fig_coer,ax_coer = plt.subplots(1,1)
ax_coer.scatter(nominal_sizes, coer)
ax_coer.set_xlabel('Nominal diameter (nm)')
ax_coer.set_ylabel('Coercive field (Oe)')
plt.show()

print(list(arrays.values())[:].properties['nominal_diameter'])
```
