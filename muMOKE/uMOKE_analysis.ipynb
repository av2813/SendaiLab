```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scienceplots
sns.set_palette('colorblind')

plt.style.use('science')
sns.set_context('talk')


def read_uMOKE_data(path):
    """Reads uMOKE data from a CSV file.

    Args:
        path (str): Path to the CSV file.

    Returns:
        tuple: A tuple containing two pandas DataFrames. The first DataFrame
               contains the main data, and the second contains the reference data.
    """
    data = pd.read_csv(path, header=31, skipfooter=2002, encoding='latin1')
    data_ref = pd.read_csv(path, header=2033, encoding='latin1')
    return data, data_ref


def plot_MOKE(data, fig=None, ax=None):
    """Plots the MOKE hysteresis loop.

    Args:
        data (pandas.DataFrame): The DataFrame containing the uMOKE data.
        fig (matplotlib.figure.Figure, optional): Existing figure to use. 
            Defaults to None.
        ax (matplotlib.axes._axes.Axes, optional): Existing axis to use. 
            Defaults to None.

    Returns:
        tuple: A tuple containing the created figure and axis (if None were provided).
    """
    if fig is None or ax is None:
        fig, ax = plt.subplots(1, 1)

    m_sig = data['Theta k'] - np.mean(data['Theta k'])
    ax.plot(data['H Signal'], m_sig, '-')
    ax.axvline(0, zorder=-10, color='k')
    ax.axhline(0, zorder=-10, color='k')
    ax.set_xlabel('H (kOe)')
    ax.set_ylabel('M (theta rot.)')
    return fig, ax


folder = r'C:\Users\alexv\OneDrive - Imperial College London\Projects\Japan2024_May_KSAV\Japan_visit\P-AOMS\uMOKE\av13524-S-dotsv2-02'

# Get all files with .csv extension
all_files = os.listdir(folder)
good_files = []

ignore_list = []
# Filter files based on desired criteria
for file in all_files:
    if file.endswith('.csv') and not any(word in file.lower() for word in ignore_list):  # Replace "ignore" and "exclude" with your unwanted word(case-insensitive)
        good_files.append(file)

fig,ax = plt.subplots(1,1)
# Loop through each file and plot the data
for file in good_files:
    filepath = os.path.join(folder, file)
    data, data_ref = read_uMOKE_data(filepath)
    fig, ax = plot_MOKE(data,fig,ax)
fig.suptitle(f"MOKE Data: S-Dotsv1-SingleLayerTa/Pt/Co/Pt", fontsize=12)  # Add file name as title
plt.show()

```
