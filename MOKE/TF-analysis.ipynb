```python
# Comparison plots 
import scienceplots
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as mcolors

def create_paired_colormap():
    # Define 8 distinct colors in pairs
    colors = [
        "#1f77b4", "#aec7e8",  # Blue pair
        "#ff7f0e", "#ffbb78",  # Orange pair
        "#2ca02c", "#98df8a",  # Green pair
        "#d62728", "#ff9896"   # Red pair
    ]
    
    return mcolors.ListedColormap(colors)
paired_cmap = create_paired_colormap()
plt.style.use(['science', 'ieee', 'no-latex'])
base_folder 
folder_list = [r'SiSend-CoPt5xTF-TaPt4Co1Pt2x5-041024',
               r'SiSend-CoPt1xTF-TaPt4Co1Pt2x1-041024',
               r'SiO2Imp-CoPt5xTF-Ta2Pt4Co1Pt2x5-041024',
               r'SiO2Imp-CoPt1xTF-Ta2Pt4Co1Pt2-041024',
               r'Au-CoPt5xTF-Ta2Pt4Co1Pt2x5-041024',
               r'Au-CoPt1xTF-Ta2Pt4Co1Pt2x1-041024',
               r'AL250-CoPt5xTf-Ta2Pt4Co1Pt2x5-041024',
               r'Al250-CoPt1xTF-Ta2Pt4Co1Pt2-041024']

summary_data_filename = r'coercive_fields_summary.csv'
sum_data_dict = {}
fig_coer,ax_coer = plt.subplots(1,1)
fig_std,ax_std = plt.subplots(1,1)
for i,folder in enumerate(folder_list):
    path = os.path.join(os.path.join(base_folder,folder), summary_data_filename)
    sum_data_dict[i] = pd.read_csv(path)
    print(sum_data_dict[0])
    color = paired_cmap(i / len(folder_list))
    ax_coer.plot(np.log(sum_data_dict[i]['sweep_rate']),sum_data_dict[i]['Hc_mean'],'-o', label=folder, color = color)
    
    ax_std.plot(np.log(sum_data_dict[i]['sweep_rate']),sum_data_dict[i]['sfd_std_mean'],'o', label=folder, color = color)
    ax_coer.set_xlabel('Rate (Oe/s) - log scale')
    ax_coer.set_ylabel('$H_{c}$ (Oe)')
    ax_std.set_xlabel('Rate (Oe/s) - log scale')
    ax_std.set_ylabel('SFD width ($Oe^{-1}$)')
    ax_std.set_ylim(0,100)
ax_std.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small')
ax_coer.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small')

plt.show()

```
