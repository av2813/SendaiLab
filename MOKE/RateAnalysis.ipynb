```python
import numpy as np
import matplotlib.pyplot as plt
import lmfit
from scipy.optimize import curve_fit
import scienceplots

plt.style.use(['science', 'ieee', 'no-latex'])

max_field = 2500
Hcs = np.array([478, 510, 543, 637, 684])   # In Oe
time = np.array([1000, 300, 100, 10, 5]) # Time in seconds
rate = 4*max_field/time
log_rate = np.log(rate)
def linear(x,a,b):
    return(a*x+b)

popt, _ = curve_fit(linear, log_rate, Hcs)

fig, ax = plt.subplots(1,1)
ax.plot(log_rate, Hcs, '.', label = 'AlSiO2-CoPt-1xTF')
ax.plot(log_rate, popt[0]*log_rate+popt[1],'-',label = 'Linear fit\n $H_c$ = {:.1f}*ln(R)+{:.1f}'.format(popt[0], popt[1]))
ax.set_xlabel('ln(Rate) (Oe/s)')
ax.set_ylabel('$H_{c}$ (Oe)')
plt.legend()
plt.show()

fig, ax = plt.subplots(1,1)
ax.plot(rate, Hcs, '.', label = 'AlSiO2-CoPt-1xTF')
#ax.plot(log_rate, popt[0]*log_rate+popt[1],'-',label = 'Linear fit\n $H_c$ = {:.1f}*ln(R)+{:.1f}'.format(popt[0], popt[1]))
ax.set_xlabel('Rate (Oe/s)')
ax.set_ylabel('$H_{c}$ (Oe)')
plt.legend()
plt.show()

print(popt)
```
